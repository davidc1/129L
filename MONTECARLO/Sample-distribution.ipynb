### Define function to sample from

```python
def Expo(x,tau):
    return tau*np.exp(-x*tau)
```

```python
Expo_v = np.vectorize(Expo)
TAU = 0.1
xvals = np.linspace(0,10,1000)
yvals = Expo_v(xvals,TAU)
```

```python
fig = plt.figure(figsize=(6,6))
plt.plot(xvals,yvals,'k--',label=r'$e^{-%.01f x}$'%TAU)
plt.xlabel('x values')
plt.ylabel('y values')
#plt.yscale('log')
plt.legend(loc=1)
plt.show()
```

### define sampling function

```python
# sample from the distribution randomly

# step 1 -> define x-range to sample over
XMIN = 0
XMAX = 10

# step 2 -> uniform random number between XMIN and XMAX
xval = np.random.random() * (XMAX-XMIN) + XMIN
print('random number is %.02f'%xval)

# step 3 -> I need to know the maximum of my function
fmax = Expo(0,TAU)
print('Max value of f(x) is %.02f'%fmax)

# step 3 -> do we keep it?
# simulate random number between 0 and fmax
r = np.random.random()*fmax
# calculate value of f(x) at x = xval
f = Expo(xval,TAU)
print('simulated random number %.02f. f(x) at %.02f is %.02f'%(r,xval,f))
# if r < f -> keep!
if (r < f):
    print('Keep value %.02f'%r)
else:
    print('skip value %.02f'%r)
```

```python
# sample from the distribution randomly
def Sample(f,XMIN,XMAX,VERBOSE=False):
    
    # did we successfully find a number from the distribution?
    RET = False
    
    while(RET == False):

        # step 2 -> uniform random number between XMIN and XMAX
        xval = np.random.random() * (XMAX-XMIN) + XMIN
        if (VERBOSE): print('random number is %.02f'%xval)

        # step 3 -> I need to know the maximum of my function
        fmax = Expo(0,TAU)
        if (VERBOSE): print('Max value of f(x) is %.02f'%fmax)

        # step 3 -> do we keep it?
        # simulate random number between 0 and fmax
        r = np.random.random()*fmax
        # calculate value of f(x) at x = xval
        f = Expo(xval,TAU)
        if (VERBOSE): print('simulated random number %.02f. f(x) at %.02f is %.02f'%(r,xval,f))
        # if r < f -> keep!
        if (r < f):
            if (VERBOSE): print('Keep value %.02f'%r)
            return xval
        else:
            if (VERBOSE): print('skip value %.02f'%r)
```

```python
Sample(Expo,0,100,VERBOSE=True)
```

```python
# collect high-stats sample from the distribution
def SampleN(f,N):
    ctr = 0
    vals = []
    while (ctr < N):
        vals.append(Sample(f,0,100,VERBOSE=False))
        ctr += 1
    return np.array(vals)
```

### Plot High-Stats Distribution

```python
fig = plt.figure(figsize=(6,6))
plt.plot(xvals,yvals,'k--')
N = 1000
sample_vals = SampleN(Expo,N)
vals,binedges = np.histogram(sample_vals,bins=np.linspace(0,10,26))
binc = 0.5*(binedges[:-1]+binedges[1:])
vals = vals.astype(float)
errs = np.sqrt(vals)
AREA = N * (binc[1]-binc[0])
vals /= AREA
errs /= AREA
plt.errorbar(binc,vals,yerr=errs,fmt='o',color='r')
plt.show()
```
