```python
from numpy.fft import fft, ifft, ifftshift, fftfreq
```

```python
from numpy.random import normal
```

```python
noise_v = normal(0,1.5,1000)

fig = plt.figure(figsize=(10,6))
plt.plot(noise_v,'r--')
plt.show()
```

```python
# Example signal
sample_rate = 1000  # Hz
t = np.linspace(0, 1, sample_rate, endpoint=False)
signal = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 12 * t)

signal = signal+noise_v

fig = plt.figure(figsize=(10,6))
plt.plot(t,signal,'k--')
plt.xlabel('time [sec]')
plt.ylabel('amplitude')
plt.show()


```

```python
# Compute FFT
fft_result = fft(signal)
magnitude = abs(fft_result)          # Complex → magnitude
freqs = fftfreq(len(signal), d=1/sample_rate)  # Frequency axis in Hz

# Keep only positive frequencies (the spectrum is mirrored)
pos_mask = freqs >= 0
freqs = freqs[pos_mask]
magnitude = magnitude[pos_mask]

# Normalize 
magnitude = magnitude / len(signal)     # Scale by N

# Plot
plt.figure(figsize=(10, 4))
plt.plot(freqs, magnitude)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("FFT Magnitude Spectrum")
plt.grid(True)
plt.tight_layout()
plt.show()
```
