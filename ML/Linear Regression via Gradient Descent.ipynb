```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fake data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Initialize parameters
m = 0.0
b = 0.0
alpha = 0.01  # learning rate

# Gradient descent
for epoch in range(1000):
    y_pred = m * x + b

    # Compute gradients
    dm = (-2/len(x)) * np.sum(x * (y - y_pred))
    db = (-2/len(x)) * np.sum(y - y_pred)

    # Update
    m -= alpha * dm
    b -= alpha * db

# Result
print("m =", m, "b =", b)

# Plot
plt.scatter(x, y)
plt.plot(x, m*x + b)
plt.show()
```

```python
# Fake data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Initialize parameters
m = 0.0
b = 0.0
alpha = 0.01  # learning rate
```

```python
fig, ax = plt.subplots()
xvals = np.linspace(1,5,100)


# Gradient descent
for epoch in range(10):
    y_pred = m * x + b

    # Compute gradients
    dm = (-2/len(x)) * np.sum(x * (y - y_pred))
    db = (-2/len(x)) * np.sum(y - y_pred)

    # Update
    m -= alpha * dm
    b -= alpha * db
    
    # plot update
    ax.cla()
    ax.plot(x,y,'ko')
    ax.plot(xvals,m*xvals+b,'k--',label=r'$y = %.02f x + %.02f$'%(m,b))
    ax.set_ylim([-6,6])
    ax.set_xlim([-6,6])
    ax.legend(loc=2)
    display(fig)
    clear_output(wait=True)
    time.sleep(1)
```

```python
# Fake data
x = np.linspace(0,5,21)
y = np.sin(x) + np.random.random(len(x))*0.1

# Initialize parameters
A = 0.99 #0.8
f = 1.03 #2.0
alpha = 0.01  # learning rate
```

```python
fig, ax = plt.subplots()
xvals = np.linspace(1,5,100)

# Gradient descent
for epoch in range(1000):
    
    y_pred = A * np.sin(f*x)

    # Compute gradients
    dA = (-2/len(x)) * np.sum(x * (y - y_pred))
    df = (-2/len(x)) * np.sum(y - y_pred)

    # Update
    A -= alpha * dA
    f -= alpha * df
    
    # plot update
    ax.cla()
    ax.plot(x,y,'ko')
    ax.plot(xvals,A*np.sin(f*xvals),'k--')#,label=r'$y = %.02f x + %.02f$'%(m,b))
    ax.set_ylim([-1.5,1.5])
    ax.set_xlim([0,6])
    #ax.legend(loc=2)
    display(fig)
    clear_output(wait=True)
    time.sleep(0.1)
```
