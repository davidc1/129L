```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Simple 2D dataset
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                           n_clusters_per_class=1, random_state=42)

model = LogisticRegression()
model.fit(X, y)

# Plot points
plt.scatter(X[:,0], X[:,1], c=y)

# Plot decision boundary
import numpy as np
x_vals = np.linspace(X[:,0].min(), X[:,0].max(), 100)
y_vals = -(model.coef_[0][0]*x_vals + model.intercept_[0]) / model.coef_[0][1]

plt.plot(x_vals, y_vals)
plt.show()
```

```python
LEARN_RATE = 0.1
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def DW(X,p,y,n_samples):
    d = np.dot(X.T,(p-y))
    return (1./n_samples) * d
def DB(X,p,y,n_samples):
    return (1./n_samples) * np.sum(p-y)
```

```python
def updateparams(w,b,dw,db):
    w -= LEARN_RATE * dw
    b -= LEARN_RATE * db
    return w,b
```

```python
def PredictProbability(X,w,b):
    z = np.dot(X, w) + b
    return sigmoid(z)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
n_samples, n_features = X_train.shape
```

```python
print(n_samples)
print(n_features)
```

```python
xmin = np.min(X[:,0])
xmax = np.max(X[:,0])
ymin = np.min(X[:,1])
ymax = np.max(X[:,1])
```

```python
import time
from IPython.display import display, clear_output
```

```python
NITER = 1000
CTR = 0

# initialize parameters
w = np.zeros(n_features)
b = 0

fig, ax = plt.subplots()
xvals = np.linspace(xmin,xmax,100)

while (CTR < NITER):
    
    CTR += 1
    
    # predict probabilities
    p = PredictProbability(X_train,w,b)
    
    # caculate gradient
    dw = DW(X_train,p,y_train,n_samples)
    db = DB(X_train,p,y_train,n_samples)
    
    w,b = updateparams(w,b,dw,db)
    
    if(CTR%30 == 0):
        ax.cla()
        ax.scatter(X[:,0], X[:,1], c=y)

        yvals = -(w[0]*xvals + b) / w[1]

        ax.plot(xvals,yvals,'k--',label='%i iteration'%CTR)
        ax.set_xlim([xmin,xmax])
        ax.set_ylim([ymin,ymax])
        ax.legend(loc=2)
        display(fig)
        clear_output(wait=True)
        time.sleep(1)
    
    #print(w,b)
```

```python
def predict(X,w,b):
        p = PredictProbability(X,w,b)
        return (p >= 0.5).astype(int)
```

```python
y_predict = predict(X_test,w,b)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy:", accuracy_score(y_test, y_predict))
```
